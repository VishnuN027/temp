Filling Missing Values in Time-Series Data Time-series data is data that follows some special trend or seasonality. Analyzing Time series data is a little bit different than normal data frames. Whenever we have time-series data, Then to deal with missing values we cannot use mean imputation techniques. Interpolation is a powerful method to fill missing values in time-series data.

```python
print('Start date in Data : ',df['Date'].min())
print('End Date in Data   : ',df['Date'].max())
```

```python
# filling missing values
df.interpolate(limit_direction="both",inplace=True)
```

```python
#check for outliers 

plt.figure(figsize=(20,40))
features = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene', 'AQI']
for i in enumerate(features): 
    ax = plt.subplot(5, 3, i[0]+1)
    sns.boxplot(df[i[1]])
    plt.xticks(rotation = 20)
```

```python
for i in ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene', 'AQI']:
    Q1=df[i].quantile(0.25)  
    Q3=df[i].quantile(0.75)    
    IQR=Q3-Q1
    low_limit=Q1-1.5*IQR
    up_limit=Q3+1.5*IQR    
    for j in df[i]:    
       if j > up_limit or j < low_limit:     
            df[i] = df[i].replace(j, np.median(df[i]))       
```

```python
#check for outliers 

plt.figure(figsize=(20,40))
features = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene', 'AQI']
for i in enumerate(features): 
    ax = plt.subplot(5, 3, i[0]+1)
    sns.boxplot(df[i[1]])
    plt.xticks(rotation = 20)
```
