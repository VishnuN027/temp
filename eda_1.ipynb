# Attrition by MaritalStatus

```python
plt.figure(figsize=(10,5), dpi=80)
sns.countplot(data=df, x='MaritalStatus', hue='Attrition', order = df['MaritalStatus'].value_counts().index, palette='seismic_r').set_title('Attrition by MaritalStatus');
```

# MonthlyIncome and jobrole Distribution

```python
fig, ax = plt.subplots(figsize = (8,8))
sns.barplot(y = 'JobRole',x = 'MonthlyIncome',hue = 'Gender',data = df,ax = ax)
```

# No.of Companies Worked

```python
plt.figure(figsize=(15,10), dpi=50)
sns.histplot(data=df, x='NumCompaniesWorked', bins=42, kde=True).set_title(' Numer of Companies Worked');
```

# Attition based RelationshipSatisfaction
